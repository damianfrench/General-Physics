```python
from lwr_model import get_naptan_data, NAPTAN_CODES, haversine_distance, distance_lat_long, START
import pandas as pd
df = get_naptan_data()
df = df[df['NaptanCode'].isin(NAPTAN_CODES)]
```

```python
import numpy as np
def cells(lat, long, start = START):
    d_actual = distance_lat_long(start[0], start[1], lat, long)

    d_hav = haversine_distance(start[0], start[1], lat, long)

    if  1.25 * d_hav < d_actual:
        d_actual = d_hav
    return d_actual/10
```

```python
df = df.sort_values(by=['Latitude'])
df
arr = np.zeros((342))
for line in zip(df[['Latitude']].values, df[['Longitude']].values):
    lat, long = line
    cell_no = cells(lat[0], long[0], START)
    arr[int(cell_no)] = 20

print(arr)
```
